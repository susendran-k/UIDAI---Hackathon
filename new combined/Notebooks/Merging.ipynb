```python
import pandas as pd
```

```python
# THE ULTIMATE CLEANING LOGIC
state_mapping = {
    'West Bangal': 'West Bengal', 'Westbengal': 'West Bengal', 'West Bengli': 'West Bengal',
    'West  Bengal': 'West Bengal', 'WEST BENGAL': 'West Bengal', 'WESTBENGAL': 'West Bengal',
    'West bengal': 'West Bengal', 'west Bengal': 'West Bengal', 'West  Bengal': 'West Bengal',
    'Orissa': 'Odisha', 'ODISHA': 'Odisha', 'odisha': 'Odisha',
    'Jammu & Kashmir': 'Jammu And Kashmir', 'Jammu and Kashmir': 'Jammu And Kashmir',
    'Andaman & Nicobar Islands': 'Andaman And Nicobar Islands',
    'Puducherry': 'Pondicherry', 'Uttaranchal': 'Uttarakhand', 'Chhatisgarh': 'Chhattisgarh',
    'Tamilnadu': 'Tamil Nadu', 'andhra pradesh': 'Andhra Pradesh',
    'Dadra & Nagar Haveli': 'Dadra And Nagar Haveli And Daman And Diu',
    'Dadra and Nagar Haveli': 'Dadra And Nagar Haveli And Daman And Diu',
    'Daman & Diu': 'Dadra And Nagar Haveli And Daman And Diu',
    'Daman and Diu': 'Dadra And Nagar Haveli And Daman And Diu',
    'The Dadra And Nagar Haveli And Daman And Diu': 'Dadra And Nagar Haveli And Daman And Diu'
}

district_fixes = {
    'Bara Banki': 'Barabanki', 'Bulandshahar': 'Bulandshahr', 'Bagpat': 'Baghpat',
    'Howrah': 'Haora', 'Hawrah': 'Haora', 'HOWRAH': 'Haora',
    'Hooghiy': 'Hooghly', 'Hugli': 'Hooghly', 'HOOGHLY': 'Haora', # Haora/Hooghly are often mixed
    'South 24 Pargana': 'South 24 Parganas', 'South 24 parganas': 'South 24 Parganas',
    'Spsr Nellore': 'Sri Potti Sriramulu Nellore', 'Nellore': 'Sri Potti Sriramulu Nellore',
    'Gondiya *': 'Gondia', 'Nandurbar *': 'Nandurbar', 'Udupi *': 'Udupi',
    'Rangareddi': 'Rangareddy', 'K.v. Rangareddy': 'Rangareddy'
}

junk = ['100000', 'Balanagar', 'Puttenahalli', 'Raja Annamalai Puram', 'Madanapalle', 'Nagpur', 'Jaipur', 'Darbhanga']

def clean_aadhar_final(df):
    # 1. Standardize Text (Same as before)
    df['state'] = df['state'].astype(str).str.strip().str.title()
    df['district'] = df['district'].astype(str).str.strip().str.title()

    # 2. Apply Mappings (Same as before)
    df['state'] = df['state'].replace(state_mapping)
    df['district'] = df['district'].replace(district_fixes)

    # 3. PINCODE CLEANING & VALIDATION
    # Convert to numeric, turn errors to NaN, then drop NaNs
    df['pincode'] = pd.to_numeric(df['pincode'], errors='coerce')

    # Filter: Keep only 6-digit pincodes between 110000 and 900000
    df = df[(df['pincode'] >= 110000) & (df['pincode'] <= 850000)]

    # 4. Filter Junk States/Districts (like '100000')
    df = df[~df['state'].isin(junk)]
    df = df[~df['district'].isin(junk)]

    # 5. AGGREGATE (Combine the split records)
    num_cols = df.select_dtypes(include=['number']).columns.tolist()
    if 'pincode' in num_cols: num_cols.remove('pincode') # Don't sum pincodes!

    return df.groupby(['date', 'state', 'district', 'pincode'])[num_cols].sum().reset_index()

# Run it!
Aadhar_df = clean_aadhar_final(Aadhar_df)
```

```python
import numpy as np
def clean_aadhar_master_final(df):
    # TEXT STANDARDIZATION
    for col in ['state', 'district']:
        df[col] = df[col].astype(str).str.strip().replace(r'\s+', ' ', regex=True).str.title()
    # STATE MAPPING
    state_map = {
        'West Bangal': 'West Bengal', 'Westbengal': 'West Bengal', 'West Bengli': 'West Bengal', 'West  Bengal': 'West Bengal', 'Orissa': 'Odisha',
        'Uttaranchal': 'Uttarakhand', 'Chhatisgarh': 'Chhattisgarh', 'Tamilnadu': 'Tamil Nadu', 'Andaman & Nicobar Islands': 'Andaman And Nicobar Islands',
        'Dadra & Nagar Haveli': 'Dadra And Nagar Haveli And Daman And Diu', 'Daman & Diu': 'Dadra And Nagar Haveli And Daman And Diu',
        'Dadra And Nagar Haveli': 'Dadra And Nagar Haveli And Daman And Diu', 'Daman And Diu': 'Dadra And Nagar Haveli And Daman And Diu',
        'Puducherry': 'Pondicherry', 'Jammu & Kashmir': 'Jammu And Kashmir'
    }
    df['state'] = df['state'].replace(state_map)
    # DISTRICT CLEANING
    df['district'] = df['district'].str.replace('*', '', regex=False).str.strip()
    dist_map = {
        'Bara Banki': 'Barabanki', 'Bulandshahar': 'Bulandshahr', 'Bagpat': 'Baghpat', 'Howrah': 'Haora', 'Hawrah': 'Haora', 'Hooghiy': 'Hooghly', 'Hugli': 'Hooghly',
        'South 24 Pargana': 'South 24 Parganas', 'South 24 Parganas': 'South 24 Parganas', 'North Twenty Four Parganas': 'North 24 Parganas',
        'South Twenty Four Parganas': 'South 24 Parganas', 'Spsr Nellore': 'Sri Potti Sriramulu Nellore', 'Nellore': 'Sri Potti Sriramulu Nellore',
        'Anantapur': 'Ananthapuramu', 'Ananthapur': 'Ananthapuramu', 'Gondiya': 'Gondia', 'Nandurbar': 'Nandurbar', 'Udupi': 'Udupi',
        'Rangareddi': 'Rangareddy', 'K.V. Rangareddy': 'Rangareddy'
    }
    df['district'] = df['district'].replace(dist_map)
    # PINCODE CLEANING & VALIDATION
    df['pincode'] = pd.to_numeric(df['pincode'], errors='coerce')
    df = df[(df['pincode'] >= 110000) & (df['pincode'] <= 850000)].copy()
    df['pincode'] = df['pincode'].astype(int)
    # REMOVING JUNK ENTRIES
    junk = ['100000', 'Balanagar', 'Puttenahalli', 'Raja Annamalai Puram', 'Madanapalle', 'Nagpur', 'Jaipur', 'Darbhanga']
    df = df[~df['state'].isin(junk)]
    df = df[~df['district'].isin(junk)]
    num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
    if 'pincode' in num_cols: num_cols.remove('pincode')
    group_cols = ['date', 'state', 'district', 'pincode']
    actual_groups = [c for c in group_cols if c in df.columns]

    return df.groupby(actual_groups)[num_cols].sum().reset_index()

# Run the Master Cleaner
Aadhar_df = clean_aadhar_master_final(Aadhar_df)

# Print Final Statistics for your PPT
print(f"✅ FINAL DATA READY")
print(f"Unique States: {Aadhar_df['state'].nunique()}")
print(f"Unique Districts: {Aadhar_df['district'].nunique()}")
print(f"Total Records: {len(Aadhar_df)}")
```
