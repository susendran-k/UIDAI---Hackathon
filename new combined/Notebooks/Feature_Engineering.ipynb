```python
import pandas as pd, numpy as np
```

```python
Aadhar_df = pd.read_csv('Aadhar_Data.csv')
```

```python
print("Merged Aadhar Data Details : ")
print("Aadhar Data's Shape : ", Aadhar_df.shape)
print("Missing Values : ", Aadhar_df.isnull().sum().sum())
print("Info: ")
print(Aadhar_df.info())
print("Duplicates : ", Aadhar_df.duplicated().sum())
```

```python
Aadhar_df['date'] = pd.to_datetime(Aadhar_df['date'])
```

```python
Aadhar_df.info()
```

```python
Aadhar_df['total_updates'] = Aadhar_df['total_biometric'] + Aadhar_df['total_demography']
Aadhar_df['total_workload'] = Aadhar_df['total_enrollment'] + Aadhar_df['total_updates']
total_national_workload = Aadhar_df['total_workload'].sum()
Aadhar_df['maturity_pct'] = (Aadhar_df['total_updates'] / (Aadhar_df['total_enrollment'] + 1) * 100).round(2)
Aadhar_df['migration_pct'] = (Aadhar_df['demo_age_17_'] / (Aadhar_df['total_enrollment'] + 1) * 100).round(2)
Aadhar_df['adult_update_pct'] = ((Aadhar_df['bio_age_17_'] + Aadhar_df['demo_age_17_']) / (Aadhar_df['age_18_greater'] + 1) * 100).round(2)
Aadhar_df['maintenance_intensity_pct'] = (Aadhar_df['total_updates'] / (Aadhar_df['total_workload'] + 1) * 100).round(2)
Aadhar_df['workload_share_pct'] = (Aadhar_df['total_workload'] / total_national_workload * 100).round(2)
Aadhar_df['biometric_backlog_vol'] = Aadhar_df['age_0_5'] - Aadhar_df['bio_age_5_17']
```

```python
Aadhar_df['date'] = pd.to_datetime(Aadhar_df['date'], format='%d-%m-%Y')
```
